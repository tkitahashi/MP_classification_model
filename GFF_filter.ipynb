```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# load the spectral data

```python
# loading the spectral data of the dry GF/F filter
dry= pd.read_csv("dry_GFF.csv", index_col=0)
dry.head()
```

```python
# loading the spectral data of the wet GF/F filter
wet= pd.read_csv("wet_GFF.csv", index_col=0)
wet.head()
```

```python
# extract the SWIR range data
dry25= dry.iloc[:, 5:273]
dry25.head()
```

```python
# extract the SWIR range data
wet25= wet.iloc[:, 5:273]
wet25.head()
```

```python
# Convert pandas dataframes into a NumPy arrays
dry_Np= dry25.values
wet_Np= wet25.values
```

```python
# extract wavelength data
wl= dry25.columns
```

# Illustrate the spectral data (Figure S3)

```python
plt.plot(wl.astype(np.float32), np.average(dry_Np, axis=0), color='black', label="Dry GF/F filter")
plt.plot(wl.astype(np.float32), np.average(wet_Np, axis=0), color='blue', label="Wet GF/F filter")

plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance')
plt.ylim([0,1])
plt.legend()
```
